```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Estilo visual de Seaborn
sns.set(style="darkgrid")

# ----------------------------
# 1. Crear dataset de ejemplo
# ----------------------------

data = {
    "Producto": ["A", "B", "C", "D", "E", "A", "B", "C", "D", "E"],
    "Ventas": [100, 200, 150, 300, 250, 120, 220, 180, 310, 260],
    "Precio": [10, 15, 12, 20, 18, 11, 16, 13, 19, 17],
    "Categoría": ["X", "Y", "X", "Z", "Y", "X", "Y", "X", "Z", "Y"],
    "Región": ["Norte", "Sur", "Norte", "Este", "Sur", "Norte", "Sur", "Norte", "Este", "Sur"]
}

df = pd.DataFrame(data)
df.head()

# ----------------------------
# 2. Gráfico de Barras: Ventas por Producto
# ----------------------------

plt.figure(figsize=(8, 6))
sns.barplot(x="Producto", y="Ventas", data=df, estimator=sum, errorbar=None)
plt.title("Ventas Totales por Producto")
plt.show()

# ----------------------------
# 3. Gráfico de Dispersión: Precio vs Ventas
# ----------------------------

plt.figure(figsize=(8, 6))
sns.scatterplot(x="Precio", y="Ventas", hue="Categoría", data=df, s=100)
plt.title("Relación entre Precio y Ventas")
plt.show()

# ----------------------------
# 4. Mapa de Calor de Correlaciones
# ----------------------------

plt.figure(figsize=(6, 4))
correlacion = df[["Ventas", "Precio"]].corr()
sns.heatmap(correlacion, annot=True, cmap="coolwarm")
plt.title("Matriz de Correlación")
plt.show()

# ----------------------------
# 5. Boxplot de Precio por Categoría
# ----------------------------

plt.figure(figsize=(8, 6))
sns.boxplot(x="Categoría", y="Precio", data=df)
plt.title("Distribución de Precios por Categoría")
plt.show()
```
